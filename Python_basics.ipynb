### Write a guessing game that asks the user to guess a secret number between 1 and 10.

- Give feedback (too high / too low)
- Use a while loop

```python
secret = 7  # secret number between 1 and 10

while True:
    guess = input("Guess a number between 1 and 10: ")

    # Validate input
    if not guess.isdigit():
        print("Please enter a valid number!")
        continue

    guess = int(guess)

    if guess < secret:
        print("Too low!")
    elif guess > secret:
        print("Too high!")
    else:
        print("Correct! You guessed it!")
        break
```

### Write a calculator function:

- Takes two numbers and an operation (+, -, *, /)
- Returns the result

```python
def calculator(a, b, operation):
    if operation == "+":
        return a + b
    elif operation == "-":
        return a - b
    elif operation == "*":
        return a * b
    elif operation == "/":
        if b == 0:
            return "Error: Division by zero"
        return a / b
    else:
        return "Invalid operation"
    
calculator(4,6,'+')
```

### Class and Object Basics
- Define a class named Student with the following attributes:
    - name
    - age
    - grade
- Create a method display_info() that prints the student's details.
- Create three student objects and call the method for each to display their details.


```python
class Student:
    
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
    
    def display_info(self):
        """Displays student details."""
        print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")


# Creating student objects
student1 = Student("Alice", 20, "A")
student2 = Student("Brian", 22, "B+")
student3 = Student("Clara", 19, "A-")

# Displaying their information
student1.display_info()
student2.display_info()
student3.display_info()
```

### Instance vs Class Methods
- Define a class named BankAccount with:
     - account_number
     - balance
- Add methods:
    - deposit(amount)
    - withdraw(amount)
    - display_balance()
- Add a class variable to store the bank name and a class method to display the bank name.
- Create two account objects and perform deposits and withdrawals.

```python
class BankAccount:
    
    bank_name = "Equity Bank"  # Class variable
    
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance
    
    def deposit(self, amount):
        """Adds money to the account."""
        self.balance += amount
        print(f"Deposited {amount}. New Balance = {self.balance}")
    
    def withdraw(self, amount):
        """Withdraws money if available."""
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New Balance = {self.balance}")
        else:
            print("Insufficient funds!")
    
    def display_balance(self):
        print(f"Account {self.account_number} Balance: {self.balance}")
    
    @classmethod
    def display_bank_name(cls):
        """Displays the bank name."""
        print(f"Bank Name: {cls.bank_name}")
```

```python
# Using the BankAccount class
BankAccount.display_bank_name()

acc1 = BankAccount(101, 500)
acc2 = BankAccount(102, 1000)

acc1.deposit(200)
acc1.withdraw(100)
acc2.deposit(500)
acc2.withdraw(200)
```

### Encapsulation
- Create a class Car with:
    - Private attributes: __make, __model, __year, __speed
- Methods to:
    - Get and set make, model, and year.
    - Increase and decrease speed.
    - Display all car details.
- Create a Car object, update its attributes, and demonstrate encapsulation.


```python
class Car:
    
    def __init__(self, make, model, year, speed=0):
        self.__make = make
        self.__model = model
        self.__year = year
        self.__speed = speed
    
    # Get make
    def get_make(self):
        return self.__make
    
    def get_model(self):
        return self.__model
    
    def get_year(self):
        return self.__year
    
    # Set make
    def set_make(self, make):
        self.__make = make
    
    def set_model(self, model):
        self.__model = model
    
    def set_year(self, year):
        self.__year = year
    
    # Speed controls
    def increase_speed(self, amount):
        self.__speed += amount
        print(f"Speed increased to {self.__speed}")
    
    def decrease_speed(self, amount):
        self.__speed = max(0, self.__speed - amount)
        print(f"Speed decreased to {self.__speed}")
    
    def display_details(self):
        print(f"Car: {self.__year} {self.__make} {self.__model}, Speed: {self.__speed} km/h")


# Demonstrating encapsulation
car = Car("Toyota", "Corolla", 2019)
car.display_details()

car.set_make("Honda")
car.set_model("Civic")
car.set_year(2022)
car.increase_speed(50)
car.decrease_speed(20)

car.display_details()
```

### Methods with Objects as Arguments
- Create a class Circle with:
    - Attributes: radius
    - Method area() and circumference().
- Create a class Cylinder that takes a Circle object and height.
- Method volume() to compute the volume using the circle's area.
- Demonstrate by creating a Circle object and passing it to the Cylinder object.

```python
import math

class Circle:
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius ** 2
    
    def circumference(self):
        return 2 * math.pi * self.radius


class Cylinder:
    """Cylinder uses a Circle object for its base."""
    
    def __init__(self, circle, height):
        self.circle = circle
        self.height = height
    
    def volume(self):
        return self.circle.area() * self.height


# Demonstration
c = Circle(7)
cyl = Cylinder(c, 10)

print("Circle Area:", c.area())
print("Cylinder Volume:", cyl.volume())
```

### Object Relationships (Aggregation)
- Create a class Author with:
    - Attributes: name, nationality
    - Method to display author info.
- Create a class Book with:
    - Attributes: title, price, and author (Author object).
    - Method to display book details including author info.
- Create one author and multiple books related to that author.

```python
class Author:
    
    def __init__(self, name, nationality):
        self.name = name
        self.nationality = nationality
    
    def display_info(self):
        print(f"Author: {self.name}, Nationality: {self.nationality}")


class Book:
    """Book containing an Author object."""
    
    def __init__(self, title, price, author):
        self.title = title
        self.price = price
        self.author = author  # Aggregation
    
    def display_book(self):
        print(f"Book: {self.title}, Price: {self.price}")
        self.author.display_info()


# Creating objects
author = Author("Chinua Achebe", "Nigerian")

book1 = Book("Things Fall Apart", 1200, author)
book2 = Book("No Longer at Ease", 1500, author)

book1.display_book()
book2.display_book()
```
